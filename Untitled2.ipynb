```python
import cv2
import numpy as np
import string
import torch
import torch.nn as nn
import timm
import matplotlib.pyplot as plt

# 1. --- Configuration (Must match training exactly) ---
IMG_W = 256
IMG_H = 32

CHARS = (
    string.digits +
    string.ascii_uppercase +
    string.ascii_lowercase
)

idx_to_char = {i + 1: c for i, c in enumerate(CHARS)}

# 2. --- Model Definition ---
class OCRNet(nn.Module):
    def __init__(self, num_chars, dropout_p=0.2):
        super().__init__()
        # Ensure out_indices=(2,) is here to keep the model ultra-lightweight
        self.backbone = timm.create_model("mobilenetv3_small_050", pretrained=False, features_only=True, out_indices=(2,))
        
        with torch.no_grad():
            dummy_input = torch.zeros(1, 3, IMG_H, IMG_W)
            dummy_features = self.backbone(dummy_input)
            out_channels = dummy_features[0].shape[1] 

        self.rnn = nn.GRU(
            input_size=out_channels,
            hidden_size=128,
            num_layers=1,
            bidirectional=True,
            batch_first=True
        )
        
        self.head = nn.Sequential(
            nn.Linear(256, 256),
            nn.Hardswish(inplace=True),
            nn.LayerNorm(256),
            nn.Dropout(dropout_p),
            nn.Linear(256, num_chars + 1)
        )

    def forward(self, x):
        features = self.backbone(x)
        x = features[0]  # Index is 0 because we sliced the backbone
        x = x.mean(dim=2)
        x = x.permute(0, 2, 1)
        x, _ = self.rnn(x)  
        x = self.head(x)  
        return x

# 3. --- Helper Functions ---
def ctc_decode(logits):
    pred = logits.argmax(-1)
    text = []
    prev = -1
    for p in pred:
        p = p.item()
        if p != prev and p != 0:
            text.append(idx_to_char[p])
        prev = p
    return "".join(text)

def predict_cropped_image(model, img, device, img_w=256, img_h=32):
    # 1. Grayscale
    if len(img.shape) == 3:
        if img.shape[2] == 4:
            gray = cv2.cvtColor(img, cv2.COLOR_BGRA2GRAY)
        else:
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    else:
        gray = img.copy()

    # 2. Soften the Stucco Texture
    # A light Gaussian blur smooths the pits in the wall so the network ignores them
    gray = cv2.GaussianBlur(gray, (3, 3), 0)
    
    # 3. Auto-Polarity Correction (On the Raw Grayscale)
    # If the background is light (median > 127), invert it!
    # Dark text on light wall -> Light text on dark wall.
    if np.median(gray) > 127:
        gray = cv2.bitwise_not(gray)

    # 4. Aspect-Ratio Resize
    h, w = gray.shape
    scale = min(img_w / w, img_h / h)
    new_w = max(1, int(w * scale))
    new_h = max(1, int(h * scale))
    
    resized = cv2.resize(gray, (new_w, new_h), interpolation=cv2.INTER_AREA)

    # 5. Pad with the exact background color (now guaranteed to be dark)
    bg_color = int(np.median(gray))
    canvas = np.full((img_h, img_w), bg_color, dtype=np.uint8)
    
    x_off = (img_w - new_w) // 2
    y_off = (img_h - new_h) // 2
    canvas[y_off:y_off+new_h, x_off:x_off+new_w] = resized

    # 6. Tensor formatting
    canvas = canvas.astype(np.float32) / 255.0
    tensor_img = np.stack([canvas, canvas, canvas], axis=0) 

    x = torch.tensor(tensor_img).unsqueeze(0).to(device)    

    model.eval()
    with torch.no_grad():
        logits = model(x)[0]
        
    return ctc_decode(logits)
    
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


print(f"Using device: {device}")
model = OCRNet(num_chars=len(CHARS)).to(device)

model.load_state_dict(torch.load("best_ocr.pth", map_location=device, weights_only=True))
print("Weights loaded successfully!")

```

```python
img = cv2.imread('/home/sachin/Downloads/photo-1494871262121-49703fd34e2b.avif')
cropped_img = 255 - img[880:1150, 1000:2300]

plt.imshow(cropped_img)
# Pass the raw crop directly to preserve the character shapes
prediction = predict_cropped_image(model, cropped_img, device)
print(f"Predicted Text: {prediction}")
```
